** In the titles column there are "Reasons/Departments" specified before the title code. These are EMS, Fire, and Traffic.

```python
data['Reason'] = data['title'].apply(lambda x: x.split(':')[0])
data.head()
```

```python
data['Reason'].value_counts()
```

```python
sns.countplot(data = data, x = 'Reason', palette= 'coolwarm')
```

___
** Now let us begin to focus on time information. What is the data type of the objects in the timeStamp column? **

```python
type(data['timeStamp'].iloc[0])
```

** You should have seen that these timestamps are still strings. 

```python
data['timeStamp'] = pd.to_datetime(data['timeStamp'])
```

```python
data['Hour'] = data['timeStamp'].apply(lambda x: x.hour)
data['Month'] = data['timeStamp'].apply(lambda x : x.month)
data['Day of Week'] = data['timeStamp'].apply(lambda x : x.dayofweek)
data.head(5)
```

** Notice how the Day of Week is an integer 0-6. Use the .map() with this dictionary to map the actual string names to the day of the week: **

    dmap = {0:'Mon',1:'Tue',2:'Wed',3:'Thu',4:'Fri',5:'Sat',6:'Sun'}

```python
dmap = {0:'Mon',1:'Tue',2:'Wed',3:'Thu',4:'Fri',5:'Sat',6:'Sun'}
```

```python
data['Day of Week'] = data['Day of Week'].map(dmap)
```

** Now use seaborn to create a countplot of the Day of Week column with the hue based off of the Reason column. **

```python
sns.countplot(x = 'Day of Week',data = data, hue = 'Reason',palette='viridis')
plt.legend(bbox_to_anchor = (1.05,1),loc = 2,borderaxespad = 0. )
```

**Now do the same for Month:**

```python
sns.countplot(x = 'Month',data = data, hue = 'Reason', palette= 'coolwarm')
plt.legend(bbox_to_anchor = (1.05,1),loc = 2,borderaxespad = 0.)
```

**Create a new column called 'Date' that contains the date from the timeStamp column. You'll need to use apply along with the .date() method. ** 

```python
data['Date'] = data['timeStamp'].apply(lambda x: x.date())
data.head(10)
```

** Now groupby this Date column with the count() aggregate and create a plot of counts of 911 calls.**

```python
data.groupby('Date').count()['twp'].plot()
plt.tight_layout()
```

** Now recreate this plot but create 3 separate plots with each plot representing a Reason for the 911 call**

```python
data[data['Reason'] == 'Traffic'].groupby('Date').count()['twp'].plot()
plt.title('Traffic')
plt.tight_layout()
```

```python
data[data['Reason']=='Fire'].groupby('Date').count()['twp'].plot()
plt.title('Fire')
plt.tight_layout()
```

```python
data[data['Reason']=='EMS'].groupby('Date').count()['twp'].plot()
plt.title('EMS')
plt.tight_layout()
```

** Now let's move on to creating  heatmaps with seaborn and our data. We'll first need to restructure the dataframe so that the columns become the Hours and the Index becomes the Day of the Week.

```python
dayHour = data.groupby(by = ['Day of Week','Hour']).count()['Reason'].unstack()
dayHour.head(5)
```

** Now create a HeatMap using this new DataFrame. **

```python
plt.figure(figsize=(12,6))
sns.heatmap(data = dayHour,cmap= 'viridis')
```

** Now create a clustermap using this DataFrame. **

```python
plt.figure(figsize = (12,6))
sns.clustermap(dayHour,cmap='viridis')
```
