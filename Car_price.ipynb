Thể hiện các giá trị tương quan giữa các biến phân loại

```python
plt.figure(figsize=(25,6))
plt.subplot(1,3,1)
plt1 = cars['cylindernumber'].value_counts().plot(kind = 'bar')
plt.title('Number of cylinders')
plt.xlabel('Number of cylinders')
plt.ylabel('Frequency of Number of cylinders')

plt.subplot(1,3,2)
plt1 = cars['fueltype'].value_counts().plot(kind = 'bar')
plt.title('Fuel type')
plt.xlabel('Fuel type')
plt.ylabel('Frequency of fuel type')

plt.subplot(1,3,3)
plt1 = cars['carbody'].value_counts().plot(kind = 'bar')
plt.title('Car body')
plt.xlabel('Car body')
plt.ylabel('Frequency of car body')
```
